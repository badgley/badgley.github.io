Good question! `header` in CSS is just an **HTML element selector** — it targets any `<header>` tag in your HTML. It's not magic, it just matches the tag name.

##### 🤖Reply🤖<!-- SOLVEIT_SEPARATOR_7f3a9b2c -->

Yeah, naming collisions with semantic HTML elements (`header`, `article`, `section`, `footer`) are a classic trap. A few rules of thumb:

1. **Never style bare HTML elements globally** if that element appears in multiple contexts. Use class selectors instead — e.g. `.site-header` rather than `header`.

2. **Use `body >` or a wrapper class** to scope top-level elements when you do style them by tag, as you just did.

3. **BEM naming** (Block-Element-Modifier) is a popular convention — e.g. `.nav__link`, `.nav__link--active`. Verbose but collision-proof.

4. **CSS custom properties** (`--color-accent: purple`) defined at `:root` keep values consistent without repetition.

For a small site like yours, the simplest approach is probably just to add a class to your top nav — change `<header>` to `<header class="site-header">` in `base.html` and scope all your nav CSS to `.site-header`.

The `.` means it's a **class selector** — it targets any HTML element that has that class applied.

So `.container` matches `<div class="container">`, and `.page` matches `<div class="page">`.

Without the dot, `article` matches the `<article>` HTML tag directly — that's called an **element selector**.

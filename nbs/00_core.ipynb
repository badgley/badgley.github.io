so why are all these CSS entries prefixed with "header" — does CSS know about the various parts of the page? what other parts are predefined?

##### 🤖Reply🤖<!-- SOLVEIT_SEPARATOR_7f3a9b2c -->

Good question! `header` in CSS is just an **HTML element selector** — it targets any `<header>` tag in your HTML. It's not magic, it just matches the tag name.

The same works for any HTML element. Common ones you'd use:

- `body` — the whole page
- `nav` — navigation
- `main` — main content area
- `footer` — bottom of page
- `h1`, `h2`, `p`, `a` — headings, paragraphs, links

So when you write `header nav a { ... }`, you're saying: *"find an `<a>` inside a `<nav>` inside a `<header>`"* — it's just nesting.

The other way to target things is with **classes** (`.brand`) and **IDs** (`#main-content`), which you define yourself.

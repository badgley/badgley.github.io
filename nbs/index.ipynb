## Usage

### Installation

Install latest from the GitHub [repository][repo]:

```sh
$ pip install git+https://github.com/badgley/badgley-blog.git
```

or from [conda][conda]

```sh
$ conda install -c badgley badgley_blog
```

or from [pypi][pypi]


```sh
$ pip install badgley_blog
```

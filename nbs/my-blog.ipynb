##### 🤖Reply🤖<!-- SOLVEIT_SEPARATOR_7f3a9b2c -->

Sure! Here's a simplified version of what's happening:

```python
@rt('/notes')
def get(): ...

@rt('/')
def get(): ...
```

The `@rt('/notes')` decorator calls something like `app.route('/notes', methods=['GET'])(get)` — it registers the function under that path immediately. By the time the second `def get()` runs, the first one is already stored in the router under `/notes`. The name `get` in your local scope just gets overwritten, but the router has already captured the first function.

So the router is essentially a dictionary mapping `(path, method)` → function. The function name doesn't matter at all after registration.

````
# | export
def get_note_paths():
    return list(Path('notes').glob('*.md'))


def previews():
    note_paths = get_note_paths()
    preview_d = {}
    for note_path in note_paths:
        note = load_note(note_path)
        preview_d[note.get('date')] = preview_entry(note) # can only have one post a day — that's fine!

    preview_d = dict(sorted(preview_d.items(), reverse=True))

    return Container(H2("Notes"), *preview_d.values(), cls=(ContainerT.xs, 'space-y-4 pt-16 w-full'))


def load_note(fn):
    with open(fn) as f:
        note = frontmatter.load(f)
        note['slug'] = Path(fn).stem
        return note


def preview_entry(note):
    return A(
        Div(
            P(Strong(note.get('title'))), 
            P(note.get('preview'), cls='!text-black'),        
        ),
        href=f'notes/{note.get("slug")}', cls='block space-y-2 text-black hover:text-purple-600 transition-colors'
    )

    
````
